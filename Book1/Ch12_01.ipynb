```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris_sns = sns.load_dataset('iris')
```

```python
iris_sns
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 9))

sns.histplot(data=iris_sns, x="sepal_length", binwidth=0.2, ax = ax1)
ax1.axvline(x = iris_sns.sepal_length.mean(), color = 'r', ls='--')
ax1.axvline(x = iris_sns.sepal_length.mean() - iris_sns.sepal_length.std(), color = 'g', ls='--')
ax1.axvline(x = iris_sns.sepal_length.mean() + iris_sns.sepal_length.std(), color = 'g', ls='--')


sns.histplot(data=iris_sns, x="sepal_width", binwidth=0.2, ax = ax2)
ax2.axvline(x = iris_sns.sepal_width.mean(), color = 'r', ls='--')
ax2.axvline(x = iris_sns.sepal_width.mean() - iris_sns.sepal_width.std(), color = 'g', ls='--')
ax2.axvline(x = iris_sns.sepal_width.mean() + iris_sns.sepal_width.std(), color = 'g', ls='--')


sns.histplot(data=iris_sns, x="petal_length", binwidth=0.2, ax = ax3)
ax3.axvline(x = iris_sns.petal_length.mean(), color = 'r', ls='--')
ax3.axvline(x = iris_sns.petal_length.mean() - iris_sns.petal_length.std(), color = 'g', ls='--')
ax3.axvline(x = iris_sns.petal_length.mean() + iris_sns.petal_length.std(), color = 'g', ls='--')


sns.histplot(data=iris_sns, x="petal_width", binwidth=0.2, ax = ax4)
ax4.axvline(x = iris_sns.petal_width.mean(), color = 'r', ls='--')
ax4.axvline(x = iris_sns.petal_width.mean() - iris_sns.petal_width.std(), color = 'g', ls='--')
ax4.axvline(x = iris_sns.petal_width.mean() + iris_sns.petal_width.std(), color = 'g', ls='--')
```

```python
fig, ax = plt.subplots(figsize=(8,6))

sns.histplot(data=iris_sns, x="sepal_length",
            hue="species", binwidth=0.2, ax=ax,
            element="step", stat="density")
```

```python
sns.histplot(iris_sns, x='sepal_length', y='species', hue='species', legend=True)
```

```python
sns.barplot(data=iris_sns, x="species", y="sepal_length", width=0.4)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.kdeplot(data=iris_sns, x='sepal_length',
           bw_adjust=0.3, fill=True)
sns.rugplot(data=iris_sns, x='sepal_length')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.kdeplot(data=iris_sns, x='sepal_length', hue='species',
           bw_adjust=0.5, fill=True)
sns.rugplot(data=iris_sns, x='sepal_length', hue='species')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.kdeplot(data=iris_sns, x='sepal_length', hue='species',
           multiple='stack', bw_adjust=0.5)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.kdeplot(data=iris_sns, x='sepal_length',
           hue='species', bw_adjust=0.5, multiple='fill')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.stripplot(data=iris_sns, x='sepal_length', y='species',
             hue='petal_length', palette='RdYlBu_r', ax = ax)
```

```python
sns.stripplot(data=iris_sns, x='sepal_length', y='species', hue='petal_length',
             jitter=False, s=20, marker='D', linewidth=1, alpha=.1, palette='RdYlBu_r')
```

```python
fig, ax = plt.subplots(figsize=(8,2))
sns.violinplot(data=iris_sns, x='sepal_length', ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

sns.violinplot(data=iris_sns, x='sepal_length',
              y='species', ax = ax)
```

```python
sns.violinplot(data=iris_sns, x='sepal_length', y='species', inner='stick')
```

```python
fig, ax = plt.subplots(figsize=(8,4))

sns.swarmplot(data=iris_sns, x='sepal_length', y='species',
             hue='species', ax=ax)
```

```python
sns.catplot(data=iris_sns, x="sepal_length", y="species", 
            kind="violin", color=".9", inner=None)

sns.swarmplot(data=iris_sns, x="sepal_length", 
              y="species", size=3)
```
