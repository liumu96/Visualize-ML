```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris_sns = sns.load_dataset('iris')
```

```python
iris_sns
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 9))

sns.histplot(data=iris_sns, x="sepal_length", binwidth=0.2, ax = ax1)
ax1.axvline(x = iris_sns.sepal_length.mean(), color = 'r', ls='--')
ax1.axvline(x = iris_sns.sepal_length.mean() - iris_sns.sepal_length.std(), color = 'g', ls='--')
ax1.axvline(x = iris_sns.sepal_length.mean() + iris_sns.sepal_length.std(), color = 'g', ls='--')


sns.histplot(data=iris_sns, x="sepal_width", binwidth=0.2, ax = ax2)
ax2.axvline(x = iris_sns.sepal_width.mean(), color = 'r', ls='--')
ax2.axvline(x = iris_sns.sepal_width.mean() - iris_sns.sepal_width.std(), color = 'g', ls='--')
ax2.axvline(x = iris_sns.sepal_width.mean() + iris_sns.sepal_width.std(), color = 'g', ls='--')


sns.histplot(data=iris_sns, x="petal_length", binwidth=0.2, ax = ax3)
ax3.axvline(x = iris_sns.petal_length.mean(), color = 'r', ls='--')
ax3.axvline(x = iris_sns.petal_length.mean() - iris_sns.petal_length.std(), color = 'g', ls='--')
ax3.axvline(x = iris_sns.petal_length.mean() + iris_sns.petal_length.std(), color = 'g', ls='--')


sns.histplot(data=iris_sns, x="petal_width", binwidth=0.2, ax = ax4)
ax4.axvline(x = iris_sns.petal_width.mean(), color = 'r', ls='--')
ax4.axvline(x = iris_sns.petal_width.mean() - iris_sns.petal_width.std(), color = 'g', ls='--')
ax4.axvline(x = iris_sns.petal_width.mean() + iris_sns.petal_width.std(), color = 'g', ls='--')
```

### åæ–¹å·®çŸ©é˜µ

```python
SIGMA = iris_sns.iloc[:, :-1].cov()

fig, axs = plt.subplots()

h = sns.heatmap(SIGMA, cmap='RdYlBu_r', lw=.05, annot=True)
h.set_aspect('equal')
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
features = list(iris_sns.columns)[:-1]

g1 = sns.heatmap(iris_sns.loc[iris_sns.species == 'setosa', features].cov(),
                cmap='RdYlBu_r', fmt='.2f',
                annot=True, cbar=False, ax=ax1, square=True,
                vmax=0.4, vmin=0)
ax1.set_title('Setota')

g2 = sns.heatmap(iris_sns.loc[iris_sns.species == 'versicolor', features].cov(),
                cmap='RdYlBu_r', fmt='.2f',
                annot=True, cbar=False, ax=ax2, square=True,
                vmax=0.4, vmin=0)
ax2.set_title('Versicolor')

g1 = sns.heatmap(iris_sns.loc[iris_sns.species == 'virginica', features].cov(),
                cmap='RdYlBu_r', fmt='.2f',
                annot=True, cbar=False, ax=ax3, square=True,
                vmax=0.4, vmin=0)
ax3.set_title('Virginica')
```
