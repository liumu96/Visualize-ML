```python
import math
import matplotlib.pyplot as plt
```

```python
radius = 2

circle_area = math.pi * radius**2
print(circle_area)

circumference = 2 * math.pi * radius
print(circumference)
```

```python
num = 100
x_start = -3
x_end = 3
step = (x_end - x_start) / (num - 1)
```

```python
x_array = [x_start + i * step for i in range(num)]
x_array
```

```python
y_array = [math.exp(x_idx) for x_idx in x_array]
```

```python
# 可视化正弦函数
plt.plot(x_array, y_array, marker = '.',
         markersize = 8,
         markerfacecolor="w",
         markeredgecolor='k')
plt.xlabel('x'); plt.ylabel('y')
plt.title(r'$y=e^{x}$')
plt.grid(True)
plt.show()
```

```python
y_array_2 = [math.exp(-x_idx**2) for x_idx in x_array]
plt.plot(x_array, y_array_2, marker = '.',
         markersize = 8,
         markerfacecolor="w",
         markeredgecolor='k')
plt.xlabel('x'); plt.ylabel('y')
plt.title(r'$y=e^{(-x^2)}$')
plt.grid(True)
plt.show()
```

利用 random 和 statistics 库函数完成如下实验。抛一枚质地均匀六面色子 1000 次，色子点 数为 [1, 2, 3, 4, 5, 6]。可视化均值随抛掷次数变化。

```python
import random
import statistics
import matplotlib.pyplot as plt
```

```python
num_flips = 1000

results = [random.choices([1, 2, 3, 4, 5, 6])[0] for _ in range(num_flips)]

running_means = [statistics.mean(results[0:idx+1]) for idx in range(num_flips)]
```

```python
plt.plot(range(1, num_flips + 1), running_means)
plt.axhline(3.5, color='r')
plt.xlabel("Number of coin flips")
plt.ylabel("Running Mean")
plt.grid(True)
plt.ylim(1, 6)
plt.show()
```

利用 random 和 statistics 库函数完成如下实验。抛一枚质地不均匀色子 1000 次。点数 [1, 2, 3, 4, 5, 6] 对应的概率分别为 [0.2, 0.16, 0.16, 0.16, 0.16, 0.16]。可视化 点数均值随抛掷次数变化。

```python
num_flips = 1000

results_2 = [random.choices([1, 2, 3, 4, 5, 6], [0.2, 0.16, 0.16, 0.16, 0.16, 0.16])[0] for _ in range(num_flips)]

running_means_2 = [statistics.mean(results_2[0:idx+1]) for idx in range(num_flips)]
```

```python
plt.plot(range(1, num_flips + 1), running_means_2)
plt.axhline(3.5, color='r')
plt.xlabel("Number of coin flips")
plt.ylabel("Running Mean")
plt.grid(True)
plt.ylim(1, 6)
plt.show()
```
