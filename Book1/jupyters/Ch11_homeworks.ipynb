Q1. 分别用 Matplotlib、Seaborn、Plotly 绘制鸢尾花数据集，花瓣长度、宽度散点图，并适当 美化图像。

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris()

petal_length = iris.data[:, 2]
petal_width = iris.data[:, 3]
target = iris.target
```

```python
fig, ax = plt.subplots()

plt.scatter(petal_length, petal_width, c=target, cmap='rainbow')

plt.title('Iris petal length vs width')
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')

ax.set_xticks(np.arange(1, 7 + 1, step=1))
ax.set_yticks(np.arange(0, 3, step=1))

ax.axis('scaled')

ax.grid(ls='--', lw=0.25, c=[0.7,0.7,0.7])

ax.set_xbound(lower=1, upper=7)
ax.set_ybound(lower=0, upper=3)

plt.show()
```

```python
import numpy as np
import plotly.express as px
```

```python
iris_df = px.data.iris()
```

```python
fig = px.scatter(iris_df, x='petal_length', y='petal_width',
                 color='species',
                width=600, height=600,
                labels={'petal_length': 'Petal length (cm)',
                        'petal_width': 'Petal width (cm)'})
fig.update_layout(xaxis_range=[1, 7], yaxis_range=[0, 3],
                 xaxis = dict(tickmode='array',
                             tickvals=np.arange(1,7+1)),
                 yaxis = dict(tickmode='array',
                             tickvals=np.arange(0,3+1)),
                 legend = dict(yanchor="top", y=0.99,
                              xanchor="left", x=0.01));
fig.show()
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots()

ax = sns.scatterplot(data=iris_df, x="petal_length", y="petal_width", hue="species")

ax.set_xlabel('Petal length (cm)')
ax.set_ylabel('Petal width (cm)')
ax.set_xticks(np.arange(1, 7 + 1, step=1))
ax.set_yticks(np.arange(0, 3 + 1, step=1))
ax.axis('scaled')
ax.grid(linestyle='--', linewidth=0.25, color=[0.7, 0.7, 0.7])
ax.set_xbound(lower = 1, upper = 7)
ax.set_ybound(lower = 0, upper = 3)
```

Q2. 分别用 Matplotlib 和 Plotly 绘制如下二元函数等高线图，并用语言描述图像特点 (等高线形 状、疏密分布、增减、最大值、最小值等等)。
$$
f(x_1, x_2) = x_1 \\
f(x_1, x_2) = x_2 \\
f(x_1, x_2) = x_1 + x_2 \\
f(x_1, x_2) = x_1 - x_2 \\
f(x_1, x_2) = x_1^2 + x_2^2 \\
f(x_1, x_2) = -x_1^2 - x_2^2 \\
f(x_1, x_2) = x_1^2 + x_2^2 + x_1x_2 \\
f(x_1, x_2) = x_1^2 - x_2^2 \\
f(x_1, x_2) = x_1^2 \\
f(x_1, x_2) = x_1x_2 \\
f(x_1, x_2) = x_1^2 + x_2^2 + 2x_1x_2 \\
$$

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))


fig, ((ax1, ax2, ax3), 
      (ax4, ax5, ax6), 
      (ax7, ax8, ax9), 
      (ax10, ax11, ax12)) = plt.subplots(4, 3, figsize=(10, 10), sharey=True)

Z = X
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax1.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = Y
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax2.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X + Y
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax3.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X - Y
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax4.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X**2 + Y**2
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax5.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = -X**2 - Y**2
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax6.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X**2 + Y**2 + X*Y
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax7.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X**2 - Y**2 
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax8.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X**2 
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax9.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X*Y 
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax10.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

Z = X**2 + Y**2 + 2*X*Y
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax11.contour(X, Y, Z, levels=levels, cmap = 'RdYlBu_r')

plt.show()
```

```python
import plotly.graph_objects as go
```

```python
levels = dict(start=-4, end=4, size=0.5)
x1_array = np.linspace(-3, 3, 121)
x2_array = np.linspace(-3, 3, 121)

xx1, xx2 = np.meshgrid(x1_array, x2_array)
ff = xx1**2 + xx2**2


data = go.Contour(x=x1_array, y=x2_array, z=ff,
                 contours_coloring='lines',
                 line_width=2,
                 colorscale='RdYlBu_r',
                 contours=levels)
layout = go.Layout(width=600,
                  height=600,
                  xaxis=dict(title=r'$\x_1$'),
                  yaxis=dict(title=r'$\x_2$'))

fig = go.Figure(data=data, layout=layout)

fig.show()
```
