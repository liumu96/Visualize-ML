```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
np.einsum('ij->i', X)
```

```python
np.einsum('ij->', X)
```

```python
n = X.shape[0]  # 样本数量
mean_X = np.einsum('ij->j', X) / n
```

```python
mean_X
```

```python
X_c = X - mean_X  # 中心化数据
variance = np.einsum('ij,ij->j', X_c, X_c) / (n - 1)
```

```python
variance
```

```python
cov_matrix = np.einsum('ij,ik->jk', X_c, X_c) / (n - 1)
```

```python
cov_matrix
```
